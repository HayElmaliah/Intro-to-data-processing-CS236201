```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import imageio
import cv2
from skimage import io, color
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from matplotlib.pyplot import figure
```

Initialize parameters

```python
A = 2500
w_x = 2
w_y = 7
x_intervals = 1024
y_intervals = 1024
bits = 8
digitized_phi = np.zeros((y_intervals,x_intervals), dtype = np.float64)
```

required functions

```python
def phi(x, y, A, w_x, w_y):
  return A  *math.cos(2*math.pi*w_x*x) * math.sin(2*math.pi*w_y*y)

def sampling(represented_signal, signal, N_x, N_y, A, w_x, w_y):
  for i in range(N_y):
    for j in range(N_x):
      represented_signal[i][j]= signal(j/(N_x-1), i/(N_y-1), A, w_x, w_y)
  return represented_signal

def uniform_quantization(sampled_signal, bits):
  # as we have seen in class:
  delta = (sampled_signal.max() - sampled_signal.min())/(2**bits)
  quantized_signal = sampled_signal.min() + (np.floor((sampled_signal - sampled_signal.min())/delta)+0.5)*delta
  return quantized_signal
```

### Q1.b - plotting the digitized function $\phi$

```python
phi_digitized = uniform_quantization(sampled_signal=sampling(represented_signal=digitized_phi, signal=phi, N_x=x_intervals, N_y=y_intervals, A=A, w_x=w_x, w_y=w_y), bits=bits)
_ = plt.imshow((phi_digitized+A)/2*A, cmap='gray', origin='lower')
```

### calculate derivatives for Q1.c

```python
def phi_horizontal_derivative(x, y, A, w_x, w_y):
  return -A*2*math.pi*w_x*math.sin(2*math.pi*w_x*x)*math.sin(2*math.pi*w_y*y)

def phi_vertical_derivative(x, y, A, w_x, w_y):
  return A*2*math.pi*w_y*math.cos(2*math.pi*w_x*x)*math.cos(2*math.pi*w_y*y)

def energy(function_digitized, N_x, N_y):
  return np.sum((function_digitized**2)/(N_x*N_y))
```

```python
phi_horizontal_deriv = np.zeros((x_intervals,y_intervals), dtype = np.float64)
phi_vertical_deriv = np.zeros((x_intervals,y_intervals), dtype = np.float64)

phi_digitized_horizontal_derivative = uniform_quantization(sampling(phi_horizontal_deriv, phi_horizontal_derivative ,x_intervals, y_intervals, A, w_x, w_y), bits)
phi_digitized_vertical_derivative = uniform_quantization(sampling(phi_vertical_deriv, phi_vertical_derivative, x_intervals, y_intervals, A, w_x, w_y), bits)

horizontal_derivative_energy = energy(phi_digitized_horizontal_derivative, x_intervals, y_intervals)
vertical_derivative_energy = energy(phi_digitized_vertical_derivative, x_intervals, y_intervals)
value_range = digitized_phi.max() - digitized_phi.min()

print('Phi value-range is: [' + str(phi_digitized.min()) + ' , ' + str(phi_digitized.max()) + ']')
print('The horizontal-derivative energy of Phi is: ' + str(horizontal_derivative_energy))
print('The vertical-derivative energy of Phi is: ' + str(vertical_derivative_energy))
```

implementing the numerical parameter calculation

```python
def optimal_parameters_calc(horizontal_energy, vertical_energy, value_range, budget):
  b = math.log2((budget*(value_range**2) * math.log(2)) / math.sqrt(horizontal_energy * vertical_energy)) / 2
  N_x = math.sqrt((budget/b) * math.sqrt(horizontal_energy / vertical_energy))
  N_y = N_x * math.sqrt(vertical_energy / horizontal_energy)
  if(round(b)*round(N_x)*round(N_y)<=budget):
    return round(N_x), round(N_y), round(b)
  return math.floor(N_x),math.floor(N_y),math.floor(b)
```

applying the search procdure on B_high and B_low (and later printing the outputs)

```python
N_x_low_opt, N_y_low_opt, b_low_opt = optimal_parameters_calc(horizontal_derivative_energy, vertical_derivative_energy, value_range, 5000)
N_x_high_opt, N_y_high_opt, b_high_opt = optimal_parameters_calc(horizontal_derivative_energy, vertical_derivative_energy, value_range, 50000)
phi_digitized_budget_low_opt = np.zeros((N_y_low_opt,N_x_low_opt), dtype = np.float64)
phi_digitized_budget_high_opt = np.zeros((N_y_high_opt, N_x_high_opt), dtype = np.float64)
print('The optimal low budget  N_x =  ' + str(N_x_low_opt) + '  N_y = ' + str(N_y_low_opt) + '  b = ' + str(b_low_opt))
print('The optimal high budget N_x =  ' + str(N_x_high_opt) + '  N_y = ' + str(N_y_high_opt) + '  b = ' + str(b_high_opt))
```

implementing the searching procedure

```python
def search_bit_allocation(horizontal_energy, vertical_energy, value_range, budget):
  min_MSE = np.float64(-1)
  optimal_coeff = []
  for b in range(1,32):
    for N_x in range(1,budget//b+1):
      N_y = budget//(N_x*b)
      range_component = (1/12)*((value_range*value_range)/(2**(2*b)))
      horizontal_energy_component = (1/(12*(N_x*N_x)))*horizontal_energy
      vertical_energy_component = (1/(12*(N_y*N_y)))*vertical_energy
      MSE = horizontal_energy_component + vertical_energy_component + range_component
      if (MSE < min_MSE or min_MSE == -1):
        min_MSE = MSE
        optimal_coeff = [N_x,N_y,b]
  return optimal_coeff
```

## Question 2 - Hadamard, Hadamard-Walsh, and Haar matrices

Q2.a.

```python
def Hadamard_matrix_calc(n):
  if n==1:
    return np.array([[1,1],[1,-1]], dtype=float) / math.sqrt(2)
  Hadamard_matrix = np.zeros((2**n,2**n),dtype=float)
  Hadamard_sub_matrix = Hadamard_matrix_calc(n-1)
  Hadamard_matrix[0:2**(n-1),0:2**(n-1)]=Hadamard_sub_matrix
  Hadamard_matrix[2**(n-1):2**(n),0:2**(n-1)]=Hadamard_sub_matrix
  Hadamard_matrix[0:2**(n-1),2**(n-1):2**(n)]=Hadamard_sub_matrix
  Hadamard_matrix[2**(n-1):2**(n),2**(n-1):2**(n)]=-Hadamard_sub_matrix
  return Hadamard_matrix / math.sqrt(2)
```

```python
def sign_changes_per_row(N, Hadamard_row):
  sign_changes = 0
  for i in range(N-1):
    if Hadamard_row[i] != Hadamard_row[i+1]:
      sign_changes += 1
  return sign_changes

def Walsh_Hadamard_matrix_calc(Hadamard):
  N = Hadamard.shape[0]
  permutation_matrix = np.zeros((N,N))
  for i in range(N):
    sign_changes = sign_changes_per_row(N, Hadamard[i])
    permutation_matrix[sign_changes,i] = 1
  Walsh_Hadamard_matrix = permutation_matrix@Hadamard
  return Walsh_Hadamard_matrix
```

```python
def Haar_matrix_calc(N):
  if N == 1:
    return Hadamard_matrix_calc(N)
  sub_Haar_matrix=Haar_matrix_calc(N-1)
  right_op=np.zeros((2,1))
  right_op[0,0]=1
  right_op[1,0]=-1
  eye=np.eye(2**(N-1))
  Haar_right_half=np.kron(eye,right_op)
  left_op=np.zeros((2,1))
  left_op[0,0]=1
  left_op[1,0]=1
  Haar_left_half=np.kron(sub_Haar_matrix,left_op)
  Haar = np.c_[Haar_left_half,Haar_right_half]
  Haar = Haar/math.sqrt(2)
  return Haar
```

```python
def basis_coeff_calc(basis):
  basis_coeff = []
  for func in basis:
    coeff = np.array(4)
    low = np.array([-4+i * delta for i in range(4)])
    high = np.array([-4+(i+1) * delta for i in range(4)])
    coeff = np.sum(func*(np.exp(high)*(high - 1) - np.exp(low)*(low-1)))
    basis_coeff.append(coeff)
  return basis_coeff
```

```python
standard = standard_basis_calc(2)/3
standard_coeff = basis_coeff_calc(standard)
k_term_standard_ordered_idx = np.argsort(np.abs(standard_coeff)).tolist()
k_term_standard_ordered_idx.reverse()
```

Calculate MSE for different bases

```python
signal_energy = (1/4) * (np.exp(2 * high_t) * (2 * high_t**2 - 2 * high_t + 1) - np.exp(2 * low_t) * (2 * low_t**2 - 2 * low_t + 1))
```

```python
bases_dict = {
    "standard": (standard_coeff, k_term_standard_ordered_idx),
    "haar": (Haar_coeff, k_term_haar_ordered_idx),
    "hadamard": (Hadamard_coeff, k_term_hadamard_ordered_idx),
    "walsh-hadamard":(Hadamard_Walsh_coeff, k_term_Hadamard_Walsh_ordered_idx)
}
```

```python
MSE_dict = {}
for basis_name, coef_idx_tuple in bases_dict.items():
  MSE_list = []
  for i in range(4):
    sum_squared_coeffs = 0
    for k in coef_idx_tuple[1][:i+1]:
        sum_squared_coeffs += np.power(coef_idx_tuple[0][k], 2)
    MSE_list.append(signal_energy - sum_squared_coeffs)
  MSE_dict[basis_name] = MSE_list
```

```python
for basis_name,MSE_basis_list in MSE_dict.items():
  plt.figure()
  k_terms = [1,2,3,4]
  plt.plot(k_terms, MSE_basis_list)
  plt.xlabel('top-k terms')
  plt.ylabel('MSE')
  plt.title(f'MSE {basis_name} basis approx for top-k terms')
  plt.grid()
  plt.show()
```
